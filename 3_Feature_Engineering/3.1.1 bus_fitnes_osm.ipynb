```python
# Функция для времени дня
def get_day_time(timestamp):
    hour = timestamp.hour
    if 6 <= hour < 10:
        return 'morning'
    elif 10 <= hour < 18:
        return 'day'
    elif 18 <= hour < 22:
        return 'evening'
    else:
        return 'night'

# Для категории дня
def get_week_category(date):
    day_type = date.weekday()
    if day_type < 5:
        return 'weekday'
    else:
        return 'weekend'
        return 'weekend'
```

```python
df_tracks_spb['start_date'] = pd.to_datetime(df_tracks_spb['start_date'], utc=True)
df_tracks_spb['start_day_time'] = df_tracks_spb['start_dttm'].apply(get_day_time)
df_tracks_spb['start_week_class'] = df_tracks_spb['start_date'].apply(get_week_category)
df_tracks_spb['week_day'] = df_tracks_spb['start_dttm'].dt.day_name()

df_tracks_spb['sec_diff'] = (df_tracks_spb['end_dttm'] - df_tracks_spb['start_dttm']).dt.total_seconds()
df_tracks_spb['hour_diff'] = round(df_tracks_spb['sec_diff'] / 3600, 2)
```

```python
df_tracks_spb_sample['start_dttm'] = pd.to_datetime(df_tracks_spb_sample['start_dttm'])
df_tracks_spb_sample['end_dttm'] = pd.to_datetime(df_tracks_spb_sample['end_dttm'])
df_tracks_spb_sample['start_date'] = pd.to_datetime(df_tracks_spb_sample['start_date'])

# Aggregation function for 'h3_9'
def aggregate_data(df):
    # Aggregations
    agg_df = df.groupby('h3_9').agg(
        visit_count=('id', 'count'),
        avg_duration=('hour_diff', 'mean'),
        weekday_visits=('start_week_class', lambda x: (x == 'weekday').sum()),
        weekend_visits=('start_week_class', lambda x: (x == 'weekend').sum()),
    ).reset_index()

    # Duration categories
    duration_categories = ['Stay0_05h', 'Stay05_1h', 'Stay1_3h', 'Stay3_6h', 'Stay6_9h', 'Stay9_15h', 'Stay15_20h', 'Stay20_plus']
    duration_bins = [0, 0.5, 1, 3, 6, 9, 15, 20, float('inf')]
    df['duration_category'] = pd.cut(df['hour_diff'], bins=duration_bins, labels=duration_categories)
    duration_counts = df.groupby(['h3_9', 'duration_category']).size().unstack(fill_value=0).reset_index()

    # Merge duration counts into the aggregated DataFrame
    agg_df = pd.merge(agg_df, duration_counts, on='h3_9', how='left')

    # Weekday and time-of-day visits
    day_visits = df.pivot_table(index='h3_9', columns='week_day', aggfunc='size', fill_value=0).reset_index()
    time_visits = df.pivot_table(index='h3_9', columns='start_day_time', aggfunc='size', fill_value=0).reset_index()

    # Merge day and time visits into the aggregated DataFrame
    agg_df = pd.merge(agg_df, day_visits, on='h3_9', how='left')
    agg_df = pd.merge(agg_df, time_visits, on='h3_9', how='left')

    return agg_df
```
