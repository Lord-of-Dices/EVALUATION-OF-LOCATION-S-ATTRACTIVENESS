## Example of the interpretation for the first case:

In our dataset, the group of visitors aged 20 and below which falls into the 60th to 80th percentile range for unique visits, appears in 50% of the cases (antecedent support = 50%).We see interactions between this age group and salary group 5 (whose unique visit share also falls within the 60th to 80th percentile range). That combination of rules (age_20_unique_60_80_prcntl + salary_5_unique_share_60_80_prcntl) occurs in approximately 11% of cases (support = 11%). The confidence of observing salary group 5's presence in this percentile range given the age group's presence is 22%. It provide 1.10 lift.

- **Antecedent Support (50%)**: tells us that the condition where the age group of 20 and under falls into the 60-80 percentile range for unique visits occurs in half of all observations. This is substantial, indicating a common scenario.

- **Consequent Support (20%)**:  tells that the share of unique visits by representatives of salary group 5 falling into the 60-80 percentile range occurs in 20% of the observations.

- **Support (11%)**: shows combined occurrence (both age and salary conditions met) is seen in 11% of the observations. SHows how often conditions co-occur.

- **Confidence (22%)**: There is a 22% probability that the salary condition is met given that the age condition is already met.

- **Lift (1.10)**: A lift greater than 1 indicates a higher occurrence of this rule than would be expected if the two sets were independent.

## Apriori with 20 percentile step for each column except salary unique columns.

**Approach presented there combines percentile column of ranges 60-80 and 80-100 into one column**. So it will be easire for apriory to track rules for rich people. 
<br>**Example: In previous case we had:**
1. 'salary_5_unique_share_0_20_prcntl',
2. 'salary_5_unique_share_20_40_prcntl',
3. 'salary_5_unique_share_40_60_prcntl',
4. 'salary_5_unique_share_60_80_prcntl',
5. 'salary_5_unique_share_80_100_prcntl'

**After algorythm we will have:**
1. 'salary_5_unique_share_0_20_prcntl',
2. 'salary_5_unique_share_20_40_prcntl',
3. 'salary_5_unique_share_40_60_prcntl',
4. 'salary_5_unique_share_60_100_prcntl'. 

So we combined salary_5_unique_share_60_80_prcntl and salary_5_unique_share_80_100_prcntl into one column

**Function to make that combination of columns**

```python
binary_df_salary_share_changed = binary_df.copy()
```

```python
def combine_salary_ranges(df, salary_levels, salary_class):

    for level in salary_levels:
        col_60_80 = f'salary_{level}_{salary_class}_share_60_80_prcntl'
        col_80_100 = f'salary_{level}_{salary_class}_share_80_100_prcntl'
        new_col = f'salary_{level}_{salary_class}_share_60_100_prcntl'

        if col_60_80 in df.columns and col_80_100 in df.columns:
            # logical OR operation
            df[new_col] = df[col_60_80] | df[col_80_100]
            df.drop([col_60_80, col_80_100], axis=1, inplace=True)
    return df
```

```python
salary_levels = [1, 2, 3, 4, 5]
salary_class = 'unique'

binary_df_salary_share_changed = combine_salary_ranges(binary_df_salary_share_changed, salary_levels, salary_class)
```

```python
salary_levels = [1, 2, 3, 4, 5]
salary_class = 'total'

binary_df_salary_share_changed = combine_salary_ranges(binary_df_salary_share_changed, salary_levels, salary_class)
```
