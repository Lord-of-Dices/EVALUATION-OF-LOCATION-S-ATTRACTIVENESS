## Apriori

**we have such a small tresholds for support and confidence because of the dataset form. Since it based on 5 equal percentile range groups, there is no features with high support rate** 

```python
%%time

frequent_itemsets = apriori(binary_df, min_support=0.1, use_colnames=True, verbose=1, max_len = 2)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

# Sort the rules by lift
rules_sorted = rules.sort_values(by='lift', ascending=False)
rules_sorted.head(10)
```

## FP Grows

**for FP Grows we need boolean dataset since it works with optimization beter**

```python
binary_df_bool = binary_df.astype(bool)
```

```python
binary_df_bool.shape
```

```python
binary_df_bool.head(3)
```

```python
%%time
from mlxtend.frequent_patterns import fpgrowth, association_rules
frequent_itemsets_fp = fpgrowth(binary_df_bool, min_support=0.1, use_colnames=True, max_len=2)

rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.3)

rules_sorted_fp = rules_fp.sort_values(by='lift', ascending=False)

rules_sorted_fp.head(10)
```
